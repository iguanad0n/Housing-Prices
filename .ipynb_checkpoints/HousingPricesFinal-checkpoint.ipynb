## Определение пропущенных значений

Функция для анализа пропущенных данных в датафрейме: показывает количество и процент пропусков по столбцам с визуальным выделением.

```python
def missing_data(df):
    missing = df.isnull().sum().sort_values(ascending=False)
    total_missing = missing[missing != 0]
    percent_missing = np.around(((missing / len(df) * 100)[(missing / len(df) * 100) != 0]), decimals=2)
    return pd.concat([total_missing, percent_missing], axis=1, keys = ['Total', 'Percent'])

missing_data(X).style.background_gradient('rocket_r')
```

```python
col_missing_drop = {'PoolQC', 'MiscFeature', 'Alley'}
```

В столбцах 'PoolQC', 'MiscFeature' и 'Alley' слишком много пропущенных значений. В дальнейшим их необходимо удалить.

# Предобработка данных

## Числовые столбцы

Определяем числовые столбцы для дальнейшей обработки.

```python
cat_num = list(set(X_train.select_dtypes(exclude='object').columns))
```

Пайплайн для обработки числовых признаков:

1. numerical_transform:
- заменяет пропущенные значения на медиану;
- применяет стандартизацию.


```python
numerical_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

## Категориальные столбцы

Определяем бинарные, порядковые и номинальные категориальные столбцы для дальнейшей обработки.

```python
cat_col_transform = X_train.select_dtypes(include='object')
cat_ordinal = list({'LotShape', 'Utilities', 'LandSlope', 'ExterQual', 'ExterCond',
                    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                    'BsmtFinType2', 'HeatingQC', 'KitchenQual', 'FireplaceQu',
                    'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
                    'Fence', 'PoolQC'} - col_corr_drop - col_missing_drop)
cat_bin = list({'Street', 'Alley', 'CentralAir'} - col_corr_drop - col_missing_drop)
cat_nominal = list(set(cat_col_transform) - set(cat_ordinal) - set(cat_bin))
```

Создаем словари для кодирования бинарных и порядковых категориальных признаков в числовые значения.

```python
CentralAir_map = {'Y': 1, 'N': 0}
Street_map = {'Pave': 1, 'Grvl': 0}
binary_mapping = [
    list(globals()[col + '_map'].keys())
    for col in cat_bin
]

LotShape_map = {'Reg': 3, 'IR1': 2, 'IR2': 1, 'IR3': 0}
Utilities_map = {'AllPub': 3, 'NoSewr': 2, 'NoSeWa': 1, 'ELO': 0}
LandSlope_map = {'Gtl': 2, 'Mod': 1, 'Sev': 0}
ExterQual_map = {'Ex': 4, 'Gd': 3, 'TA': 2, 'Fa': 1, 'Po': 0}
ExterCond_map = {'Ex': 4, 'Gd': 3, 'TA': 2, 'Fa': 1, 'Po': 0}
BsmtQual_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
BsmtCond_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
BsmtExposure_map = {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1, 'NA': 0}
BsmtFinType1_map = {'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'NA': 0}
BsmtFinType2_map = {'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'NA': 0}
HeatingQC_map = {'Ex': 4, 'Gd': 3, 'TA': 2, 'Fa': 1, 'Po': 0}
KitchenQual_map = {'Ex': 4, 'Gd': 3, 'TA': 2, 'Fa': 1, 'Po': 0}
FireplaceQu_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
GarageFinish_map = {'Fin': 3, 'RFn': 2, 'Unf': 1, 'NA': 0}
GarageQual_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
GarageCond_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
PavedDrive_map = {'Y': 2, 'P': 1, 'N': 0}
Fence_map = {'GdPrv': 4, 'MnPrv': 3, 'GdWo': 2, 'MnWw': 1, 'NA': 0}
ordinal_mapping = [
    list(globals()[col + '_map'].keys())
    for col in cat_ordinal
]
```

Пайплайны для обработки различных типов категориальных признаков:

1. nominal_transform — для номинативных (категориальных без порядка):
- заменяет пропущенные значения на наиболее частые;
- применяет OneHot-кодирование (каждая категория в отдельную колонку).

2. ordinal_transform — для порядковых признаков (с естественным порядком значений):
- заменяет пропущенные значения на наиболее частые;
- преобразует категории в числовые значения согласно заданному порядку (ordinal_mapping).

3. binary_transform — для бинарных признаков (две категории):
- заменяет пропущенные значения на наиболее частые;
- кодирует значения как 0 и 1 на основе заданных словарей (binary_mapping).

```python
ordinal_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=ordinal_mapping))
])

binary_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=binary_mapping))
])

nominal_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

## Пайплайн предобработки данных

Создаем препроцессор, который применяет разные трансформации к числовым, номинальным, порядковым и бинарным признакам.

```python
preprocessor = ColumnTransformer([
    ('num', numerical_transform, cat_num),
    ('nom', nominal_transform, cat_nominal),
    ('ord', ordinal_transform, cat_ordinal),
    ('bin', binary_transform, cat_bin)
])
```

# Моделирование

Обучаем модель XGBoost с подбором гиперпараметров с помощью GridSearchCV и оцениваем её качество на валидационной выборке.

```python
model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', XGBRegressor())
])

param_grid = {
    'model__learning_rate' : [0.01, 0.1, 0.3, 0.5],
    'model__n_estimators' : [10, 50, 100, 200],
    'model__max_depth' : [3, 6, 9]
}

grid_search = GridSearchCV(full_pipeline, param_grid, cv=5, scoring='neg_root_mean_squared_error')
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
preds = best_model.predict(X_valid)

print('MAE:', mean_absolute_error(y_valid, preds))
```
